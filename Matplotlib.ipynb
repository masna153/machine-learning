```python
#How do you create a twin y-axis (secondary y-axis) in Matplotlib?
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
```

```python
#How do you create shared x-axis or y-axis for multiple subplots?
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)
```

```python
#How do you embed a smaller inset plot within a larger figure?
fig, ax = plt.subplots()
ax.plot(np.random.rand(10))
ax_inset = ax.inset_axes([0.5, 0.5, 0.4, 0.4])
```

```python
#How do you create a plot with a custom style (e.g., changing the background, font, or grid lines)?
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.plot(np.random.rand(10))
```

```python
#How do you use annotations to highlight specific points in a plot?
fig, ax = plt.subplots()
ax.plot(np.random.rand(10))
```

```python
#How do you rotate x-axis labels to improve readability in Matplotlib?
```

```python
x = np.arange(10)
y = np.random.rand(10)
labels = [f"Label {i}" for i in range(10)]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xticks(x)
ax.set_xticklabels(labels)

plt.xticks(rotation=45)
```

```python
#How do you create a plot with multiple line styles, markers, and colors dynamically?
df=pd.DataFrame(np.random.rand(10, 4), columns=['A', 'B', 'C', 'D'])
df.plot(style=['r-', 'g--', 'b:', 'k-.'])
```

```python
#How do you display a color gradient background in a Matplotlib figure?
fig, ax = plt.subplots()
ax.imshow(np.random.rand(10, 10), cmap='viridis')
```

```python
#How do you efficiently plot millions of data points using Matplotlib?
x = np.linspace(0, 10, 1000000)
y = np.sin(x)
plt.plot(x, y)
```

```python
#How do you add a trendline (regression line) to a scatter plot?
x = np.random.rand(100)
y = 2*x + np.random.rand(100)
plt.scatter(x, y)
plt.plot(x, 2*x, color='red')
```

```python
#How do you create a heatmap using Matplotlib without using Seaborn?
fig, ax = plt.subplots()
ax.imshow(np.random.rand(10, 10), cmap='viridis')
```

```python
#How do you use Matplotlib with Pandas to visualize grouped data?
df = pd.DataFrame({'A': np.random.rand(10), 'B': np.random.rand(10), 'C': np.random.rand(10)})
df.plot(kind='bar')
```

```python
#How do you create a pie chart with percentage labels in Matplotlib?
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
```

```python
#How do you create a stacked bar chart using Matplotlib?
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
df.plot(kind='bar', stacked=True)
```

```python
#How do you create a box plot and modify its whiskers, outliers, and colors?
df = pd.DataFrame({'A': np.random.rand(10), 'B': np.random.rand(10), 'C': np.random.rand(10)})
df.plot(kind='box')
```

```python
#How do you visualize a confusion matrix using Matplotlib?
from sklearn.metrics import confusion_matrix
y_true = [1, 0, 1, 1, 0, 1, 0, 0, 0, 1]
y_pred = [1, 0, 1, 0, 0, 1, 1, 1, 0, 0]
cm = confusion_matrix(y_true, y_pred)
fig, ax = plt.subplots()
ax.imshow(cm, cmap='Blues')
```
